## Информация о датасете

* `hotel` - Тип отеля: `City Hotel` - городской отель, `Resort Hotel` - курортный отель
* `is_canceled` - Была ли бронь отменена: `1` = отменена, `0` = не отменена
* `lead_time` - Время между бронированием и заездом
* `arrival_date_year` - Год заезда
* `arrival_date_month` - Месяц заезда
* `arrival_date_week_number` - Номер недели заезда
* `arrival_date_day_of_month` - День месяца заезда
* `stays_in_weekend_nights` - Количество ночей на выходных
* `stays_in_week_nights` - Количество ночей в будние дни
* `adults` - Число взрослых гостей
* `children` - Число детей
* `babies` - Число младенцев

* `meal` - Тип питания: например, коды: `BB` = Bed & Breakfast (включен завтрак), `HB` = Half Board (полупансион), `FB` = Full Board (полный пансион), `SC` = Self Catering (питание не включено)

* `country` - Страна гостя: код страны, обычно в формате ISO-кода
* `market_segment` - Сегмент/источник бронирования: например, коды: `TA` = Travel Agent, `TO` = Tour Operator и др.

* `distribution_channel` - Канал распределения брони: например, коды: TA = Travel Agent, TO = Tour Operator

* `is_repeated_guest` - Повторный ли это гость: 1 = да, 0 = нет
* `previous_cancellations` - Сколько предыдущих броней отменялось  
* `previous_bookings_not_canceled` - Сколько предыдущих броней не отменялось  
* `reserved_room_type` - Зарезервированный тип номера: его код, например A, B, C и т. д.
* `assigned_room_type` - Фактически назначенный тип номера (аналогично `reserved_room_type`)

* `booking_changes` - Количество изменений брони  
* `deposit_type` - Тип депозита: `No Deposit` = без депозита, `Non Refund` = невозвратный, `Refundable` = возвратный
* `agent` - Код агента: идентификатор турагентства/агента
* `company` - Код компании: идентификатор компании, если бронь корпоративная
* `days_in_waiting_list` - Дней в списке ожидания  
* `customer_type` - Тип клиента: например, `Transient` = разовый клиент, `Transient-Party` = группа разовых клиентов, `Contract` = контрактный клиент, `Group` = групповая бронь
* `adr` - Average Daily Rate, средняя дневная ставка
* `required_car_parking_spaces` - Нужное число парковочных мест  
* `total_of_special_requests` - Общее число специальных запросов  
* `reservation_status` - Статус брони
* `reservation_status_date` - Дата статуса брони

####Работа с пропусками

Посчитаю долю нанов от общего количества объектов для каждого столбца.

```python
def nan_counts(df):
  nans = []
  for c in df.columns:
    nans.append(len(df[pd.isna(df[c])]))

  res = zip(list(df.columns), nans)
  return pd.DataFrame(res, columns = ['column', 'nan_count']).sort_values('nan_count', ascending=False)
```

```python
nan_counts(df).head(5)
```

###Работа с типами признаков в датасете

1. БИНАРНЫЕ ПРИЗНАКИ (0 или 1)
* `is_canceled` - бронь отменена или нет.

* `is_repeated_guest` - повторный гость или нет.

2. КАТЕГОРИАЛЬНЫЕ ПРИЗНАКИ

* `hotel` - тип отеля (City Hotel / Resort Hotel).

* `arrival_date_month` - месяц заезда (название месяца).

* `meal` - тип питания (BB, HB, FB, SC).

* `country` - ISO-код страны.

* `market_segment` - сегмент бронирования (TA, TO, Direct и т.д.).

* `distribution_channel` - канал дистрибуции (TA, TO, GDS и т.д.).

* `reserved_room_type` - забронированный тип номера (коды A, B, C...).

* `assigned_room_type` - фактически выданный тип номера.

* `deposit_type` - тип депозита (No Deposit, Non Refund, Refundable).

* `customer_type` - тип клиента (Transient, Group и т.д.).

* `reservation_status` - статус брони (Canceled, Check-Out, No-Show).

* `agent` - ID агента.

* `company` - ID компании.

3. КОЛИЧЕСТВЕННЫЕ ПРИЗНАКИ

Вещественные:

* `adr` - средняя цена за сутки.

Целочисленные:

* `lead_time` - количество дней между бронью и заездом.

* `stays_in_weekend_nights` - количество ночей.

* `stays_in_week_nights` - количество ночей.

* `adults` / `children` / `babies` - количество человек.

* `previous_cancellations` - количество отмен.

* `previous_bookings_not_canceled` - количество успешных броней.

* `booking_changes` - количество изменений в бронировании.

* `days_in_waiting_list` - количество дней в очереди.

* `total_of_special_requests` - количество спец. запросов.

4. ВРЕМЕННЫЕ/ПОРЯДКОВЫЕ ПРИЗНАКИ:

* `arrival_date_year` - год заезда.

* `arrival_date_week_number` - порядковый номер недели в году.

* `arrival_date_day_of_month` - день месяца.

* `reservation_status_date` - дата изменения статуса.

Приведем колонки к соответствующим им типам данных (колонки `company` и `agent` трогать не будем).

```python
cat_cols = [
    'hotel', 'arrival_date_month', 'meal', 'country',
    'market_segment', 'distribution_channel', 'reserved_room_type',
    'assigned_room_type', 'deposit_type', 'customer_type', 'reservation_status'
]

int_cols = [
    'lead_time', 'arrival_date_year', 'arrival_date_week_number',
    'arrival_date_day_of_month', 'stays_in_weekend_nights', 'stays_in_week_nights',
    'adults', 'children', 'babies', 'previous_cancellations',
    'previous_bookings_not_canceled', 'booking_changes', 'days_in_waiting_list',
    'total_of_special_requests', 'is_canceled', 'is_repeated_guest'
]

float_cols = ['adr']

date_cols = ['reservation_status_date']
```

```python
for cat_col in cat_cols:
  df[cat_col] = df[cat_col].astype('category')

for int_col in int_cols:
  df[int_col] = df[int_col].astype('int')

for float_col in float_cols:
  df[float_col] = df[float_col].astype('float')
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```
